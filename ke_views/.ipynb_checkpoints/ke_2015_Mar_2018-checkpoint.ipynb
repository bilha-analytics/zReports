```python
## Load dataset 
db = pd.read_csv( "%s/LG/1804_KE_Annual_Profile_Data_2015-March_2018.csv" % DATA_DIR)
```

```python
## Remove HQ data 
db = db[ db["branch_name"] != "HQ"]

#db["branch_name"].unique()
```

```python
## Subset annual status data = observation is a year, records is values as at end of each year/period
yearly_db = db[ (( (db["month"]==12) & (db["year"]!=2018) ) | ( (db["month"]==3) & (db["year"]==2018) ) ) ]
yearly_db["IsActive_branch"] = np.where( yearly_db["active_chws_in_period"] > 0 , 1, 0)
#yearly_db.sample(7).T
```

```python
## argz

fig_size = ( 10, 7)

## Variables of Interest
argz_EPMz = ["active_chws_in_period", "community_events", 
             "assess_child", "total_malaria",  
             "anc_visits", "pregnancies_registered", "total_deliveries_at_hf"]

argz_program_size = ["active_chws_in_period", "families_registered", "total_assessments"]
func_program_size = {   "active_chws_in_period": { "# Active CHWs": np.sum}, 
                        "families_registered": { "# Households" : np.sum }, 
                        "total_assessments": { "# Assessments": np.sum} 
                       }

argz_ke_program_size = ["IsActive_branch"] + argz_program_size 
func_ke_program_size = dict( {"IsActive_branch": { "# Branches": np.sum}, } , **func_program_size) #zCountUnique
```

```python
t = pd.pivot_table( yearly_db, index=["branch_name"], columns=["year"], values=["active_chws_in_period"], aggfunc=np.sum)

t.iloc[0,1] 

len( t)

for i in range(len( t )):
    #print( "at row %i = %s " % ( i, t.iloc[i,:]))
    for b in range( len( t.iloc[i,:] ) ):
        #print( "%d, %s at year %d, %d = %d" % (i, t.index[i], b, t.columns[b][1], t.iloc[i,b]) )
        if( t.iloc[i,b] > 0): 
            yearly_db[ np.where( (yearly_db["branch_name"]== t.index[i]) & ( yearly_db["year"] == t.columns[b][1]) )]["IsActive_branch"] = 1


#t
pd.pivot_table(yearly_db, index=["branch_name"], columns=["year"], values=["IsActive_branch"], aggfunc=np.sum )
```
