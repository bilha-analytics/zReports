```python
pd.pivot_table( db, index=["Year"], values=["Pct.Coverage"], 
               aggfunc=np.mean ).sort_values(by=('Pct.Coverage')).plot( kind="line", figsize=(13,7), title="Immunization Coverage ")
```

```python
print( "\n\n Numbers as at 2013")
pd.pivot_table( db[ db["Year"]=="2013"], index=["County"], values=["Pct.Coverage"], 
              aggfunc=sum).sort_values(by="Pct.Coverage").plot(kind="bar", figsize=(13,7), title="Immunization Coverage in 2013")
```
