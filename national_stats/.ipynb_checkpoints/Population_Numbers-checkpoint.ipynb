```python
## Load dataset
db = pd.read_csv( "%s/KNBS_et_al/population-distribution-by-sex-number-of-households-area-and-density.csv" % DATA_DIR) 

db.head()
db.tail()
```

```python
# wrangle 
```

```python
## Compute

## Pct Females
db["Pct_Females"] = np.nan_to_num( db["Females"]/db["Total Population"])
db["Household_Size"] = np.nan_to_num( db["Total Population"]/db["No of HouseHolds"])

db.sample( 7)
```

# Introduction

```python
pd.pivot_table( db, index=["County"], values=["Household_Size"], 
               aggfunc=sum ).sort_values(("Household_Size")).plot( kind="bar", figsize=(13,7),table=False, title="Average Household Size By County - 2013/5")
```

```python
pd.pivot_table( db, index=["County"], values=["Density"], 
               aggfunc=sum ).sort_values(("Density")).plot( kind="bar", figsize=(13,7), title="County Population Density - 2013/5")
```

```python
pd.pivot_table( db, index=["County"], values=["Household_Size", "Density"], 
               aggfunc=zSumLog ).sort_values(["Density", "Household_Size"]).plot( kind="bar", figsize=(13,7), title="Average Household Size By County - 2013/5")
```

## Density Quintiles 

```python
quintilize( db, "Density")
db.head()
```

```python
pd.pivot_table( db, index=["Quints_Density"], values=["Household_Size", "Density", "Pct_Females"], 
               aggfunc=zMeanLog ).sort_values(("Density")).plot( 
    kind="bar", figsize=(13,7),table=False, title="Averages By Density Quintiles - 2013/5")
```

```python
db.groupby("Quints_Density")["Pct_Females"].mean().round(2).plot(kind="pie", table=True, title="Average % Female By Density Quintiles")
```
